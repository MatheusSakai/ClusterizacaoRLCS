```python
# Imports

import pandas as pd

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Dataframe da tabela estatisticas

df_estatisticas = pd.read_csv('Dados/estatisticasV28.csv', sep=';')
```

## Tratando valores ausentes

Aplicamos a interpolação, ou seja, preenchemos os valores ausentes com outro valor. Em algumas colunas utilizamos a mediana da coluna para o preenchimento, e em outras usamos a média. 

```python
# Colunas a serem preenchidas com a mediana da coluna
colunas_com_mediana = [
    'stt_blocos_grandes_coletados',
    'stt_blocos_grandes_roubados',
    'stt_blocos_pequenos_coletados',
    'stt_blocos_pequenos_roubados',
    'stt_demolicoes_causadas',
    'stt_demolicoes_recebidas'
]

for coluna in colunas_com_mediana:
    mediana = df_estatisticas[coluna].median()
    df_estatisticas[coluna].fillna(mediana, inplace=True)


# Preenchendo o resto dos valores ausentes com a média da coluna
colunas_restantes_com_valores_ausentes = df_estatisticas.columns[df_estatisticas.isna().any()]

for coluna in colunas_restantes_com_valores_ausentes:
    media = df_estatisticas[coluna].mean()
    df_estatisticas[coluna].fillna(media, inplace=True)
    df_estatisticas[coluna] = df_estatisticas[coluna].round(2)

```

## Pré-Processamento dos dados

```python
# Criando novo dataframe com a média das estatísticas de cada jogador

# Removemos a coluna stt_jog_id já que os ID's não devem ser usados em operações aritméticas
df_media_por_jogador = df_estatisticas.drop('stt_jog_id', axis=1)

df_media_por_jogador = df_media_por_jogador.groupby('stt_jgd_id').mean()

df_media_por_jogador = df_media_por_jogador.reset_index()

# Arredondando tudo para duas casas decimais
for coluna in df_media_por_jogador.columns:
    df_media_por_jogador[coluna] = df_media_por_jogador[coluna].round(2)

df_media_por_jogador.shape
```

```python
df_media_por_jogador.head(10)
```

```python
# Excluindo a coluna de ID já que não deve ser utilizada em operações aritméticas

df_media_por_jogador_sem_id = df_media_por_jogador.drop('stt_jgd_id', axis=1)

df_media_por_jogador_sem_id.shape
```

```python
# Criando padronizador de dados

padronizador = StandardScaler()
```

```python
# Aplicando o padronizador nas colunas do dataframe sem as colunas de ID's

dados_padronizados = padronizador.fit_transform(df_media_por_jogador_sem_id)
```

```python
# Lembrando que a informação do dado continua a mesma, porém em outra escala

print(dados_padronizados)
```

## Construindo modelo de machine learning para segmentação de jogadores usando KMeans

```python
# Número de clusters (k)

k = 3
```

```python
# Criando o modelo kmeans

modelo_kmeans = KMeans(n_clusters=k)
```

```python
# Treinando o modelo usando os dados padronizados

modelo_kmeans.fit(dados_padronizados)
```

```python
# Adicionando a coluna cluster

df_media_por_jogador['cluster'] = modelo_kmeans.labels_
```
